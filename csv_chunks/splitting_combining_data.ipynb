Loads the single file, splits it evenly, and downloads it so we can upload to github. Will need the file path and probably the indexes updated depending on who uses it/what changes have been made.

```python
df1 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender1.csv')
df2 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender2.csv')
df3 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender3.csv')
df4 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender4.csv')
df5 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender5.csv')
df6 = pd.read_csv('~/Desktop/school/info/490/dblp_with_gender6.csv')

df1 = df1.append(df2, ignore_index = True)
df1 = df1.append(df3, ignore_index = True)
df1 = df1.append(df4, ignore_index = True)
df1 = df1.append(df5, ignore_index = True)
df1 = df1.append(df6, ignore_index = True)

df = df1

df.to_csv('dblp_with_gender.csv', encoding='utf-8', index=False)
```
