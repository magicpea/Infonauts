```python
import pandas as pd
import numpy as np
```

```python
col_names = pd.read_csv('~/Desktop/school/info/490/dblp-to-csv-master/output_article_header.csv',sep=';')
# column names are in a separate csv
df = pd.read_csv('~/Desktop/school/info/490/Infonauts/csv_chunks/csv10.csv',names=list(col_names))
```

```python
df = df[['article:ID',
'author:string[]',
'ee:string[]',
'journal:string',
'key:string',
'mdate:date',
'pages:string',
'title:string',
'url:string[]',
'volume:string',
'year:int']]
df = df[df['article:ID'] != 0]
```

```python
df
```

```python
import gender_guesser.detector as gender
d = gender.Detector(case_sensitive=False)


female_count = []
mostly_female_count = []
andy_count = []
mostly_male_count = []
male_count = []
unknown_count = []
gender_col = []
for i in range(0, len(df)):
    female_c = 0
    mostly_female_c = 0
    andy_c = 0
    mostly_male_c = 0
    male_c = 0
    unknown_c = 0
    genders = []
    full_names = str(df['author:string[]'].iloc[i]).split('|')
    for name in full_names:
        first_name = str(name).split(" ")[0]
        if d.get_gender(first_name) == 'female':
            female_c += 1
        if d.get_gender(first_name) == 'mostly_female':
            mostly_female_c += 1
        if d.get_gender(first_name) == 'andy':
            andy_c += 1
        if d.get_gender(first_name) == 'mostly_male':
            mostly_male_c += 1
        if d.get_gender(first_name) == 'male':
            male_c += 1
        if d.get_gender(first_name) == 'unknown':
            unknown_c += 1
        genders.append(d.get_gender(first_name))
    female_count.append(female_c)
    mostly_female_count.append(mostly_female_c)
    andy_count.append(andy_c)
    mostly_male_count.append(mostly_male_c)
    male_count.append(male_c)
    unknown_count.append(unknown_c)
    gender_col.append(genders)
```

```python
female_count
```

```python
df["gender_list"] = gender_col
df["female_count"] = female_count
df["mostly_female_count"] = mostly_female_count
df["andy_count"] = andy_count
df["mostly_male_count"] = mostly_female_count
df["male_count"] = male_count
df["unknown_count"] = unknown_count
df
```

```python
df.to_csv('chunk_10_genders.csv', encoding='utf-8', index=False)
```
